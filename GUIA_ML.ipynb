```python
# Capa de una red neuronal
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Forward pass
x = np.array([2, 3])    # Entrada
W = np.array([[0.5, 0.1], [0.2, 0.8]])  # Pesos (2x2)
b = np.array([0.1, 0.2])                # Bias
z = W @ x + b           # Salida capa lineal
a = sigmoid(z)          # Activacin

# Supongamos que dL/da = [-0.1, 0.2] (gradiente de la prdida respecto a 'a')
dL_da = np.array([-0.1, 0.2])

# Regla de cadena: dL/dW = dL/da * da/dz * dz/dW
da_dz = a * (1 - a)     # Derivada de sigmoid: '(z) = (z)(1 - (z))
dz_dW = x               # Derivada de z respecto a W (es la entrada x)

dL_dz = dL_da * da_dz   # dL/dz = dL/da * da/dz
dL_dW = np.outer(dL_dz, x)  # Producto exterior para dL/dW

print("dL/dW:\n", dL_dW)
```

MARKETING DIGITAL

caso: empresas quieren asignar un presupuesto de Google ads, Facebook e instagram para maximizar conversiones(ventas,lead)
gradientes para ajustar el presupuesto diario en cada plataforma en base al ROI

```python
import numpy as np

# Parmetros
cpc = np.array([0.5, 0.7, 0.3])             # Coste por clic
tasa_conversion = np.array([0.02, 0.015, 0.025])  # Conversiones por clic
presupuesto_total = 1000                   # USD/da

# Funcin objetivo: conversiones totales
def conversiones(presupuestos):
    return np.dot(presupuestos / cpc, tasa_conversion)

# Gradiente (constante en este caso)
def gradiente_conversiones():
    return tasa_conversion / cpc

# Inicializacin
presupuestos = np.array([300, 300, 400], dtype=float)
learning_rate = 10
min_presupuesto = 50
max_presupuesto = 800

# Gradient ascent con normalizacin
for _ in range(100):
    grad = gradiente_conversiones()
    presupuestos += learning_rate * grad

    # Respetar lmites
    presupuestos = np.clip(presupuestos, min_presupuesto, max_presupuesto)

    # Normalizar para que la suma sea igual al presupuesto total
    factor = presupuesto_total / np.sum(presupuestos)
    presupuestos *= factor

print("Presupuesto ptimo por plataforma:", presupuestos.round(2))
print("Conversiones mximas estimadas:", conversiones(presupuestos).round(2))
```

Caf que se Enfra
Caso: Quieres saber cmo cambia tu satisfaccin con el tiempo mientras tu caf se enfra.

Paso 1: Temperatura del caf
Paso 2: Satisfaccin
Objetivo

```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# EJEMPLO COMPLEJO: TEMPERATURA DE CAF  SATISFACCIN DEL CLIENTE
# ============================================================
# Imagina que tienes un caf y modelas lo siguiente:
# 1. La temperatura (T) de tu caf disminuye con el tiempo (t).
# 2. La satisfaccin (S) del cliente depende de la temperatura del caf.

# ------------------------------------------------------------
# PASO 1: Definir las relaciones (funciones)
# ------------------------------------------------------------
# Funcin 1: Temperatura del caf en funcin del tiempo (t en minutos)
# T(t) = 70 * e^(-0.1 * t)
def temperatura(t):
    """Calcula la temperatura del caf (en C) en un tiempo 't' dado."""
    return 70 * np.exp(-0.1 * t)

# Funcin 2: Satisfaccin del cliente en funcin de la temperatura (T en C)
# S(T) = ln(T + 1)
def satisfaccion(T):
    """Calcula la satisfaccin del cliente basada en la temperatura 'T' del caf."""
    return np.log(T + 1)

# ------------------------------------------------------------
# PASO 2: Calcular las Derivadas de las relaciones bsicas
# ------------------------------------------------------------
# Queremos saber: Cmo cambia la satisfaccin con el tiempo?
# Esto es: d(Satisfaccin)/d(tiempo) o dS/dt

# Derivada 1: Cmo cambia la temperatura con el tiempo? (dT/dt)
# T(t) = 70 * e^(-0.1 * t)
# d(T)/d(t) = 70 * (-0.1) * e^(-0.1 * t) = -7 * e^(-0.1 * t)
def dT_dt(t):
    """Calcula la tasa de cambio de la temperatura con respecto al tiempo."""
    return -7 * np.exp(-0.1 * t)

# Derivada 2: Cmo cambia la satisfaccin con la temperatura? (dS/dT)
# S(T) = ln(T + 1)
# d(S)/d(T) = 1 / (T + 1)
def dS_dT(T):
    """Calcula la tasa de cambio de la satisfaccin con respecto a la temperatura."""
    return 1 / (T + 1)

# ------------------------------------------------------------
# PASO 3: Aplicar la Regla de la Cadena (dS/dt = dS/dT * dT/dt)
# ------------------------------------------------------------
# Para calcular el cambio de satisfaccin con el tiempo, necesitamos el valor de la temperatura en ese momento.
def dS_dt_cadena(t):
    """Calcula la tasa de cambio de la satisfaccin con respecto al tiempo usando la regla de la cadena."""
    T_en_t = temperatura(t)      # Obtenemos la temperatura en el tiempo 't'
    ds_dt_val = dS_dT(T_en_t) * dT_dt(t) # Aplicamos la regla de la cadena
    return ds_dt_val

# ------------------------------------------------------------
# PASO 4: Resultado y explicacin para un tiempo especfico
# ------------------------------------------------------------
t_ejemplo = 5 # minutos
T_en_t_ejemplo = temperatura(t_ejemplo)
S_en_t_ejemplo = satisfaccion(T_en_t_ejemplo)
cambio_S_dt_ejemplo = dS_dt_cadena(t_ejemplo)

print("\n" + "="*60)
print(" ANALISIS: Tiempo  Temperatura de Caf  Satisfaccin")
print("="*60)
print(f"Para el tiempo t = {t_ejemplo} minutos:")
print(f"   Temperatura del caf (T): {T_en_t_ejemplo:.2f} C")
print(f"   Satisfaccin del cliente (S): {S_en_t_ejemplo:.2f} puntos")

print("\n" + "="*60)
print(" REGLA DE CADENA APLICADA (en t=5 min)")
print("="*60)

# Calcula los componentes individuales de la derivada en t=5 min
dT_dt_val = dT_dt(t_ejemplo)
dS_dT_val = dS_dT(T_en_t_ejemplo)

print(f"1. d(Temperatura)/d(tiempo) [dT/dt]: {dT_dt_val:.4f} C/min")
print(f"   (La temperatura est disminuyendo a esta tasa)")
print(f"2. d(Satisfaccin)/d(Temperatura) [dS/dT]: {dS_dT_val:.4f} puntos/C")
print(f"   (La satisfaccin cambia esta cantidad por cada grado de cambio en la temperatura)")
print(f"3. TOTAL d(Satisfaccin)/d(tiempo) [dS/dt]: {dS_dT_val:.4f}  {dT_dt_val:.4f} = {cambio_S_dt_ejemplo:.4f} puntos/min")
print(f"   (La satisfaccin disminuye a esta tasa en t={t_ejemplo} min debido a la cada de temperatura)")

# ------------------------------------------------------------
# PASO 5: Visualizacin (Opcional, pero til)
# ------------------------------------------------------------
tiempos = np.linspace(0, 30, 100)
temperaturas_vals = temperatura(tiempos)
satisfacciones_vals = satisfaccion(temperaturas_vals)
dS_dt_vals = dS_dt_cadena(tiempos) # Usamos la funcin de la regla de cadena

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(tiempos, temperaturas_vals, label='Temperatura T(t)')
plt.xlabel('Tiempo (min)')
plt.ylabel('Temperatura (C)')
plt.title('Temperatura del Caf vs. Tiempo')
plt.grid(True)
plt.axvline(x=t_ejemplo, color='gray', linestyle='--', label=f't={t_ejemplo} min')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(tiempos, satisfacciones_vals, label='Satisfaccin S(t)', color='orange')
plt.xlabel('Tiempo (min)')
plt.ylabel('Satisfaccin (puntos)')
plt.title('Satisfaccin del Cliente vs. Tiempo')
plt.grid(True)
plt.axvline(x=t_ejemplo, color='gray', linestyle='--', label=f't={t_ejemplo} min')
plt.legend()

plt.tight_layout()
plt.show()

# Tambin podemos graficar la tasa de cambio de satisfaccin
plt.figure(figsize=(8, 4))
plt.plot(tiempos, dS_dt_vals, label='dS/dt', color='purple')
plt.xlabel('Tiempo (min)')
plt.ylabel('Cambio de Satisfaccin (puntos/min)')
plt.title('Tasa de Cambio de Satisfaccin con el Tiempo')
plt.grid(True)
plt.axvline(x=t_ejemplo, color='gray', linestyle='--', label=f't={t_ejemplo} min')
plt.legend()
plt.show()
```

Calcula como afecta el clima en las ventas de helado

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ============================================================
# EJEMPLO ML: TEMPERATURA EXTERIOR  VENTAS DE HELADOS
# ============================================================
# Imagina que tienes una heladera y has recopilado datos para entender
# cmo la temperatura exterior (T_ext) impacta directamente en tus ventas (V).
# Esta relacin la modelas usando Regresin Lineal.

# ------------------------------------------------------------
# PASO 1: Definir la relacin (Funcin aprendida de un modelo ML)
# ------------------------------------------------------------

# Datos de ejemplo (reemplaza con datos reales de tu heladera)
temperaturas_historicas = np.array([28, 30, 25, 32, 22, 27]).reshape(-1, 1) # Temperaturas en C
ventas_historicas = np.array([45, 50, 40, 55, 35, 42]) # Ventas de Helados en unidades

# Entrenar modelo de regresin lineal
modelo_ventas_temp = LinearRegression()
modelo_ventas_temp.fit(temperaturas_historicas, ventas_historicas)

# Obtener los coeficientes del modelo de regresin
intercepto_ventas = modelo_ventas_temp.intercept_
coeficiente_ventas = modelo_ventas_temp.coef_[0] # Este es dV/dT_ext

print("\n" + "="*50)
print(" MODELO APRENDIDO: Temperatura Exterior  Ventas")
print("="*50)
print(f"La relacin aprendida es: Ventas = {intercepto_ventas:.2f} + {coeficiente_ventas:.2f} * Temperatura_Exterior")

# Definimos la funcin de ventas ajustada con los coeficientes aprendidos
def ventas_desde_temperatura(temperatura_ext):
    """
    Calcula las ventas de helados basadas en la temperatura exterior,
    usando el modelo de regresin lineal entrenado.
    """
    return intercepto_ventas + coeficiente_ventas * temperatura_ext

# Verificacin con tu dato de ejemplo:
temp_verificacion = 30
print(f"Prediccin de ventas a {temp_verificacion}C (usando modelo): {ventas_desde_temperatura(temp_verificacion):.2f} helados")

# ------------------------------------------------------------
# PASO 2: Calcular la Derivada (directamente del modelo ML)
# ------------------------------------------------------------
# Queremos saber: Cmo un cambio en la Temperatura Exterior afecta las Ventas?
# Esto es: d(Ventas)/d(Temperatura_Exterior) o dV/dT_ext

# Para un modelo de regresin lineal simple Y = a + bX,
# la derivada dY/dX es simplemente el coeficiente 'b'.

# Derivada: d(Ventas)/d(Temperatura_Exterior) = coeficiente_ventas
def dV_dT_ext():
    """Retorna la tasa de cambio de Ventas con respecto a la Temperatura Exterior (coeficiente del modelo)."""
    return coeficiente_ventas

# ------------------------------------------------------------
# PASO 3: Resultado y explicacin de la relacin directa
# ------------------------------------------------------------
cambio_ventas_por_grado = dV_dT_ext()

print("\n" + "="*50)
print(" ANALISIS DE LA RELACIN APRENDIDA")
print("="*50)
print(f"La tasa de cambio de Ventas con respecto a la Temperatura Exterior (dV/dT_ext) es:")
print(f"  {cambio_ventas_por_grado:.2f} helados/C")
print("   (Este valor es el coeficiente de la Temperatura Exterior en nuestro modelo de regresin lineal.)")
print("   (Significa que por cada grado Celsius que aumenta la temperatura exterior,")
print(f"   esperamos vender {cambio_ventas_por_grado:.2f} helados adicionales.)")

# ------------------------------------------------------------
# PASO 4: Ejemplo prctico
# ------------------------------------------------------------
temperatura_actual = 28 # C
print("\n" + "="*50)
print(f" PREDICCIN PARA HOY ({temperatura_actual}C)")
print("="*50)

ventas_esperadas_hoy = ventas_desde_temperatura(temperatura_actual)
print(f" Ventas estimadas para hoy: {ventas_esperadas_hoy:.2f} helados")

# Efecto de 1C extra:
print("\n" + "="*50)
print(" EFECTO DE 1C EXTRA EN TEMPERATURA EXTERIOR")
print("="*50)
ventas_extra_1grado = cambio_ventas_por_grado * 1
print(f" Si la temperatura aumenta 1C adicional, se venderan: {ventas_extra_1grado:.2f} helados extra")
print(f" Nuevo total de ventas estimado (si la temperatura fuera {temperatura_actual + 1}C): {ventas_esperadas_hoy + ventas_extra_1grado:.2f} helados")
```

Backpropagation

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
learning_rate = 0.05
epoch = 25

#datos de entrnamiento y = 2x
X = np.array([0, 1, 2, 3, 4], dtype=np.float32)
Y_true = np.array([0, 2, 4, 6, 8], dtype=np.float32)

#inicializacion de paramteros (peso y sesgo)
w = np.random.randn()
b = np.random.randn()

#almacenar historicos para graficas
history = {'loss': [], 'w': [], 'b': []}

#entrenamiento con backpropagation
for epoch in range(epoch):
  Y_pred = w * X + b
  loss = np.mean((Y_pred-Y_true)**2)
  #backpropagation
  der_w = np.mean (2*(Y_pred- Y_true)* X)
  der_b = np.mean(2*(Y_pred- Y_true))

  #actualizacion de parametros
  w -= learning_rate * der_w
  b -= learning_rate * der_b

  #almacenar historicos para graficas
  history['loss'].append(loss)
  history['w'].append(w)
  history['b'].append(b)

  if epoch % 5 == 0:
    print(f'Epoch: {epoch}, Loss: {loss}, w: {w:.4f}, b: {b:.4f}')

#resultados
print("\nResultados finales:")
print(f"w: {w:.4f}, b: {b:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuracin optimizada para mostrar "elbows" claros
np.random.seed(42)
LEARNING_RATE = 0.15  # Tasa de aprendizaje reducida
EPOCHS = 30           # Ms pocas para ver convergencia completa

# Datos de entrenamiento
X = np.array([0, 1, 2, 3, 4])
Y_true = 2 * X  # y = 2x

# Inicializacin con valores que generen convergencia gradual
w = 8.0  # Lejos del ptimo (2)
b = 5.0  # Lejos del ptimo (0)

# Historial detallado
history = {
    'loss': [],
    'w': [],
    'b': [],
    'epoch': []
}

# Entrenamiento con monitoreo detallado
for epoch in range(EPOCHS):
    # Forward pass
    Y_pred = w * X + b

    # Clculo de prdida (MSE)
    loss = np.mean((Y_pred - Y_true)**2)

    # Backpropagation
    dL_dw = np.mean(2 * (Y_pred - Y_true) * X)
    dL_db = np.mean(2 * (Y_pred - Y_true))

    # Actualizacin de parmetros
    w -= LEARNING_RATE * dL_dw
    b -= LEARNING_RATE * dL_db

    # Guardar histrico
    history['loss'].append(loss)
    history['w'].append(w)
    history['b'].append(b)
    history['epoch'].append(epoch)
```

## BackProgation para tienda minorista
Sistema de ventas para minoristas
una cadena de supermercados quiere predecir las vetnas diarias de leche para optimizar tus inventarios y reducir desperdicios

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# Configuracin reproducible
np.random.seed(42)

# Datos sintticos mejorados con patrones ms realistas (60 das)
dias = np.arange(1, 61)
temperatura = np.random.normal(25, 5, 60)  # Media 25C, desviacin 5
dia_semana = np.tile([1,2,3,4,5,6,7], 9)[:60]  # Secuencia de das de semana
promocion = np.where((dia_semana == 6) | (dia_semana == 7),
                     np.random.choice([0,1], p=[0.3,0.7], size=60),
                     np.random.choice([0,1], p=[0.7,0.3], size=60))
ventas_base = 100 + 20*np.sin(dias/7)  # Tendencia base semanal
ventas = ventas_base + 30*promocion + 2*temperatura + np.random.normal(0, 10, 60)

# Preparamos los datos (usamos ventas de das anteriores)
X = np.column_stack((
    temperatura[1:],
    dia_semana[1:],
    promocion[1:],
    ventas[:-1]  # Ventas del da anterior
))
y = ventas[1:]  # Ventas del da actual

# Normalizacin de caractersticas
scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()

# Hiperparmetros optimizados
LEARNING_RATE = 0.05
EPOCHS = 1000
HIDDEN_NEURONS = 5  # Red ms simple para evitar sobreajuste

# Inicializacin de pesos corregida
W1 = np.random.randn(X_scaled.shape[1], HIDDEN_NEURONS) * 0.1
b1 = np.zeros(HIDDEN_NEURONS)
W2 = np.random.randn(HIDDEN_NEURONS) * 0.1
b2 = 0.0

# Historial para visualizacin
loss_history = []
r2_history = []

# Entrenamiento con Backpropagation corregido
for epoch in range(EPOCHS):
    # Forward pass
    hidden_layer = np.tanh(np.dot(X_scaled, W1) + b1)  # Tanh para mejor convergencia
    y_pred = np.dot(hidden_layer, W2) + b2

    # Clculo de prdida
    loss = np.mean((y_pred - y_scaled)**2)
    loss_history.append(loss)

    # Calcular R peridicamente
    if epoch % 50 == 0:
        y_pred_inv = scaler_y.inverse_transform(y_pred.reshape(-1, 1))
        r2 = r2_score(y, y_pred_inv)
        r2_history.append(r2)

    # Backpropagation - CORRECCIN DE DIMENSIONES
    grad_y_pred = 2 * (y_pred - y_scaled) / len(y_scaled)

    # Gradientes capa salida
    grad_W2 = np.dot(hidden_layer.T, grad_y_pred)
    grad_b2 = np.sum(grad_y_pred)

    # Gradientes capa oculta
    grad_hidden = np.outer(grad_y_pred, W2) * (1 - hidden_layer**2)  # Derivada de tanh
    grad_W1 = np.dot(X_scaled.T, grad_hidden)
    grad_b1 = np.sum(grad_hidden, axis=0)

    # Actualizacin de pesos
    W1 -= LEARNING_RATE * grad_W1
    b1 -= LEARNING_RATE * grad_b1
    W2 -= LEARNING_RATE * grad_W2
    b2 -= LEARNING_RATE * grad_b2

# Prediccin final
y_pred_final = scaler_y.inverse_transform(
    (np.dot(np.tanh(np.dot(X_scaled, W1) + b1), W2) + b2).reshape(-1, 1)
)
```

```python
plt.figure(figsize=(16, 10))

# 1. Evolucin de la prdida
plt.subplot(2, 2, 1)
plt.plot(loss_history)
plt.title('Evolucin de la Prdida durante el Entrenamiento')
plt.xlabel('poca')
plt.ylabel('Error (MSE)')
plt.yscale('log')
plt.grid(True)

# 2. Calidad de prediccin
plt.subplot(2, 2, 2)
plt.plot(y, 'bo-', label='Ventas Reales')
plt.plot(y_pred_final, 'ro--', label='Predicciones')
plt.title('Prediccin de Ventas Diarias')
plt.xlabel('Da')
plt.ylabel('Ventas (unidades)')
plt.legend()
plt.grid(True)

# 3. Evolucin del R
plt.subplot(2, 2, 3)
plt.plot(np.arange(0, EPOCHS, 50), r2_history, 'g-')
plt.title('Evolucin del Coeficiente R')
plt.xlabel('poca')
plt.ylabel('R')
plt.ylim(0, 1)
plt.grid(True)

# 4. Anlisis de residuos
residuos = y - y_pred_final.flatten()
plt.subplot(2, 2, 4)
plt.scatter(y_pred_final, residuos, alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Anlisis de Residuos')
plt.xlabel('Predicciones')
plt.ylabel('Residuos (Real - Prediccin)')
plt.grid(True)

plt.tight_layout()
plt.savefig('retail_sales_forecasting_corrected.png', dpi=300)
plt.show()

# Mtricas finales
mape = np.mean(np.abs(residuos) / y) * 100
print(f"\nResultados del Modelo:")
print(f"- Error porcentual absoluto medio (MAPE): {mape:.2f}%")
print(f"- Coeficiente R final: {r2_history[-1]:.4f}")

# Prediccin para prxima semana
prox_semana = np.array([
    [28, 1, 0, y[-1]],  # Lunes
    [30, 2, 0, 0],       # Martes (valor dummy, se reemplaza)
    [29, 3, 0, 0],       # Mircoles
    [26, 4, 1, 0],       # Jueves (promocin)
    [24, 5, 1, 0],       # Viernes (promocin)
    [22, 6, 1, 0],       # Sbado (promocin)
    [25, 7, 0, 0]        # Domingo
])

# Usamos predicciones en cadena
predicciones = []
ultima_venta = y[-1]
for i in range(7):
    prox_semana[i, 3] = ultima_venta
    x = scaler_X.transform(prox_semana[i].reshape(1, -1))
    h = np.tanh(np.dot(x, W1) + b1)
    pred = np.dot(h, W2) + b2
    pred_real = scaler_y.inverse_transform(pred.reshape(1, -1))[0][0]
    predicciones.append(pred_real)
    ultima_venta = pred_real

print("\nPronstico para la prxima semana:")
dias_semana = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']
for dia, venta in zip(dias_semana, predicciones):
    print(f"{dia}: {venta:.0f} unidades")
```

## Regresion Lineal

##  Qu es?  
Mtodo estadstico que:  
 Encuentra relaciones LINEALES entre variables  
 Predice valores futuros basado en patrones histricos  
 Cuantifica cmo una variable afecta a otra  

**Frmula clave**:  
`y = mx + b`  
- `y`: Variable dependiente (lo que queremos predecir)  
- `x`: Variable independiente (nuestra pista)  
- `m`: Pendiente (fuerza de la relacin)  
- `b`: Intercepto (valor base cuando x=0)  

##  Elementos Clave  
| Componente | Explicacin | Ejemplo Helados |  
|------------|-------------|-----------------|  
| **Variable Independiente (x)** | Factor que controlamos o observamos | Temperatura ambiente |  
| **Variable Dependiente (y)** | Resultado que queremos predecir | Ventas de helados |  
| **Pendiente (m)** | Cunto cambia y por cada unidad de x | +5 helados/C |  
| **Intercepto (b)** | Valor base cuando x=0 | 10 helados a 0C |

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Horas de Estudio (X) y Calificaciones (Y)
horas_estudio = np.array([5, 8, 3, 10, 6, 4, 9, 7, 2, 11]).reshape(-1, 1)  # X
calificacion_examen = np.array([70, 85, 60, 90, 75, 65, 88, 80, 55, 92])   # Y

# ------------------------------------------------------------
# PASO 2: Entrenar modelo de regresin lineal
# ------------------------------------------------------------
modelo = LinearRegression()
modelo.fit(horas_estudio, calificacion_examen)

# Coeficientes del modelo
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

print("="*50)
print(" MODELO ENTRENADO")
print("="*50)
print(f" Pendiente (coef): {pendiente:.2f}")
print(f" Intercepto: {intercepto:.2f}")
print(f" Ecuacin del modelo: Y = {pendiente:.2f} * X + {intercepto:.2f}")

# ------------------------------------------------------------
# PASO 3: Predicciones
# ------------------------------------------------------------
horas_nuevas = np.array([[5], [7], [12]])  # Nuevas horas de estudio
predicciones = modelo.predict(horas_nuevas)

for h, p in zip(horas_nuevas, predicciones):
    print(f" Si estudias {h[0]} horas  Calificacin esperada: {p:.1f}")

# ------------------------------------------------------------
# PASO 4: Visualizacin
# ------------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.scatter(horas_estudio, calificacion_examen, color='blue', label="Datos reales")
plt.plot(horas_estudio, modelo.predict(horas_estudio), color='red', label="Lnea de regresin")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificacin del examen")
plt.title(" Regresin Lineal: Estudio vs Calificacin")
plt.legend()
plt.grid(True)
plt.show()
```

## Regresion Logistica

##  Qu es?  
Algoritmo de **clasificacin** (no regresin) que:  
 Predice probabilidades entre 0 y 1  
 Decide entre 2 categoras (s/no, 1/0)  
 Usa una curva en "S" (sigmoide) en lugar de lnea recta  

**Frmula clave**:  
`P(y=1) = 1 / (1 + e^(-z))`  
Donde `z = b + w1*x1 + w2*x2 + ...`  

##  Para qu se usa? (Casos prcticos)  
| Problema                   | Ejemplo Real                  |  
|----------------------------|-------------------------------|  
| Diagnstico mdico         | Tiene el paciente diabetes? |  
| Marketing digital          | El usuario har clic?       |  
| Riesgo crediticio          | Aprobar prstamo?           |  
| Deteccin de spam          | Es correo legtimo?         |  
| Calidad de produccin      | Producto defectuoso?        |  

##  Cmo funciona?  
1. Calcula una "puntuacin de riesgo" (z)  
2. Transforma a probabilidad con funcin sigmoide  
3. Clasifica segn umbral (tpicamente 0.5):  
   - **P  0.5**  Clase 1 (S)  
   - **P < 0.5**  Clase 0 (No)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# ---------------------------------------------
#  Datos simulados
# ---------------------------------------------
np.random.seed(42)
n = 50
X = np.random.rand(n, 4) * 10  # 4 caractersticas
# Generamos una relacin lineal con algo de ruido
coef = np.array([2.5, 0.0, 1.5, 0.0])  # algunas variables no importan
y = X @ coef + np.random.randn(n) * 2

# ---------------------------------------------
#  Escalamos las variables (buena prctica)
# ---------------------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------------------------------------------
#  Divisin entrenamiento/prueba
# ---------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)

# ---------------------------------------------
#  Modelos: Ridge y Lasso
# ---------------------------------------------
ridge = Ridge(alpha=1.0)
lasso = Lasso(alpha=0.1)

ridge.fit(X_train, y_train)
lasso.fit(X_train, y_train)

# ---------------------------------------------
#  Resultados
# ---------------------------------------------
print("Coeficientes Ridge:", ridge.coef_.round(2))
print("Coeficientes Lasso:", lasso.coef_.round(2))

# ---------------------------------------------
#  Visualizacin
# ---------------------------------------------
plt.figure(figsize=(10,5))
plt.bar(np.arange(4)-0.2, ridge.coef_, width=0.4, label="Ridge", alpha=0.7)
plt.bar(np.arange(4)+0.2, lasso.coef_, width=0.4, label="Lasso", alpha=0.7)
plt.xticks(np.arange(4), ['Estudio', 'Sueo', 'Tareas', 'Clases'])
plt.axhline(0, color='gray', linestyle='--')
plt.title("Coeficientes aprendidos")
plt.legend()
plt.show()
```

## Arboles de Decision

##  **Qu es?**  
Algoritmo de **ML supervisado** que:  
 Crea reglas de decisin en forma de rbol (como un flujograma)  
 Funciona para **clasificacin** (categoras) y **regresin** (valores numricos)  
 Explicable: Las decisiones son fciles de interpretar visualmente  

**Analoga**:  
"Como jugar a 20 preguntas: hace preguntas binarias (S/No) hasta llegar a una respuesta"  

##  **Cmo funciona?**  
1. **Divide** los datos en grupos homogneos usando caractersticas (features)  
2. **Elige** la mejor pregunta en cada paso (maximizando ganancia de informacin)  
3. **Repite** hasta cumplir criterios (profundidad mxima o nodos puros)

##  Ventajas vs Desventajas

|  **Ventajas**                          |  **Desventajas**                     |
|-----------------------------------------|----------------------------------------|
| Fcil de entender y visualizar          | Propenso a sobreajuste (overfitting)   |
| Maneja datos numricos y categricos    | Inestable con pequeos cambios en datos|
| No requiere normalizacin de datos      | Puede crear rboles muy profundos      |
| til para seleccin de features         | Sesgado con clases desbalanceadas      |

###  **Parmetros Principales**

| Parmetro          | Aumentar                                  | Disminuir                                   |
|--------------------|--------------------------------------------|---------------------------------------------|

| **bootstrap** (Muestreo con reemplazo) | + Mayor diversidad entre rboles <br> + Mejor estimacin de OOB (out-of-bag) | - rboles ms correlacionados <br> - Menor capacidad de generalizacin |

Predecir si un cliente pagara un prestamo o entrara en mora

```python

import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# ======================
#  DATOS SIMULADOS
# ======================
np.random.seed(42)
datos = pd.DataFrame({
    'Ingresos': np.random.randint(20000, 100000, 100),
    'Score': np.random.randint(300, 850, 100),
    'Monto': np.random.randint(5000, 50000, 100),
    'Pag': np.random.choice([0, 1], 100, p=[0.3, 0.7])  # 70% paga, 30% no
})

X = datos[['Ingresos', 'Score', 'Monto']]
y = datos['Pag']

# ======================
#  ENTRENAR MODELO
# ======================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

modelo_rf = RandomForestClassifier(
    n_estimators=100,  # Nmero de rboles
    max_depth=None,      # Profundidad mxima por rbol
    random_state=42
)
modelo_rf.fit(X_train, y_train)

# ======================
#  PREDICCIONES
# ======================
nuevos_clientes = pd.DataFrame({
    'Ingresos': [45000, 75000],
    'Score': [600, 720],
    'Monto': [20000, 35000]
})

predicciones = modelo_rf.predict(nuevos_clientes)
probabilidades = modelo_rf.predict_proba(nuevos_clientes)[:, 1]

print("Predicciones para nuevos clientes:")
for i, (pred, prob) in enumerate(zip(predicciones, probabilidades)):
    print(f"Cliente {i+1}: {' Paga' if pred == 1 else ' Mora'} ({prob:.0%} de probabilidad)")
```

### SVM "El guardian de Fronteras de ML"

##  **Qu es?**  
Algoritmo de **clasificacin/regresin** que:  
 Encuentra el **hiperplano ptimo** que separa clases  
 Usa **vectores de soporte** (puntos crticos cerca del margen)  
 Funciona con **datos lineales y no lineales** (usando kernels)  

**Analoga**:  
"Como encontrar la avenida ms ancha entre dos barrios para separarlos claramente"

##  **Cmo funciona?**  
SVM busca el **hiperplano ptimo** (lmite de decisin) que separa clases con el **mayor margen posible**. Usa **vectores de soporte** (puntos clave cerca del lmite) para definir este margen. Para datos no lineales, aplica el **kernel trick**: transforma los datos a un espacio dimensional superior donde s son separables linealmente.
##  Conceptos Clave
| Trmino | Explicacin | Visualizacin |  
|---------|------------|--------------|  
| **Hiperplano** | Superficie de decisin que separa clases | `  Hiperplano  ` |  
| **Margen** | Zona de seguridad alrededor del hiperplano | ` |Margen| ` |  
| **Vectores Soporte** | Puntos crticos que definen el margen | `   = vector soporte  ` |  
| **Kernel Trick** | Truco matemtico para separar datos no lineales | `  (Transformacin)   Separables ` |  

##  Parmetros Clave
###  Principales
| Parmetro | Aumentar  | Disminuir  |  
|----------|------------|-------------|  
| **C** (Regularizacin) | + Menos errores<br>- Margen ms pequeo<br>- Riesgo overfitting | + Ms tolerancia a errores<br>- Margen ms grande<br>- Posible underfitting |   
| **gamma** (RBF) | + Fronteras complejas<br>- Influencia local<br>- Riesgo overfitting | + Fronteras suaves<br>- Influencia global<br>- Riesgo underfitting |

###  Secundarios
| Parmetro | Aumentar  | Disminuir  |  
|----------|------------|-------------|  
| **degree** (kernel poly) | + Curvas complejas<br>- Sobreajuste | + Fronteras simples<br>- Subajuste |  
| **coef0** (poly/sigmoid) | + Ms trminos alto grado | + Ms trminos bajo grado |

### XGBOOST

##  Qu es XGBoost?

XGBoost es un modelo de Machine Learning muy poderoso y rpido que se usa para **resolver problemas de prediccin**.

- Se basa en **rboles de decisin**.
- Usa una tcnica llamada **Boosting**, que significa que varios rboles se entrenan uno despus del otro, y cada nuevo rbol trata de **corregir los errores** del rbol anterior.
- XGBoost es famoso porque es **rpido, preciso** y **gana muchas competencias de Machine Learning**, como en Kaggle.

---

##  Cmo funciona XGBoost? (versin simple)

1. El modelo comienza entrenando un rbol que hace una prediccin.
2. Despus, revisa en qu partes se equivoc.
3. Entrena otro rbol que intenta corregir esos errores.
4. Repite el proceso muchas veces, mejorando paso a paso.

As, se construyen **muchos rboles pequeos** que trabajan juntos para dar una prediccin ms precisa.

---

##  Conceptos Bsicos

| Concepto | Significado |
|---------|-------------|
| **rbol de decisin** | Modelo que toma decisiones con preguntas tipo "X es mayor que Y?". |
| **Boosting** | Tcnica que combina varios modelos simples para crear uno fuerte. |
| **Aprendizaje por etapas** | Cada rbol mejora lo que hizo mal el anterior. |
| **Regularizacin** | Mtodo para evitar que el modelo se aprenda todo de memoria (overfitting). |
| **Tasa de aprendizaje** | Cunto avanza el modelo en cada paso. Pasos pequeos = ms precisin. |

---

##  Parmetros importantes (en palabras simples)

| Parmetro | Para qu sirve? |
|-----------|------------------|
| `n_estimators` | Cuntos rboles va a crear. |
| `learning_rate` | Qu tan rpido aprende. Si es muy alto, puede cometer errores. |
| `max_depth` | Qu tan profundo (complejo) es cada rbol. |
| `subsample` | Qu parte de los datos usa cada rbol (ayuda a evitar que se sobreentrene). |
| `colsample_bytree` | Qu parte de las columnas usa cada rbol. |
| `reg_alpha` y `reg_lambda` | Ayudan a evitar que el modelo sea demasiado complejo. |
| `gamma` | Solo divide un nodo si realmente mejora el resultado. |

---

##  Ventajas (por qu usarlo?)

- Es **rpido** y funciona muy bien.
- Puede manejar datos faltantes (valores vacos).
- Suele ganar muchas competencias de prediccin.
- No necesita que transformes mucho tus datos.

---

##  Desventajas

- Puede ser difcil de entender al principio.
- Tiene muchos parmetros que hay que ajustar bien.
- Si no se controla, puede **memorizar** demasiado (overfitting).
- No es la mejor opcin para imgenes o texto directamente (hay mejores modelos para eso).


##  Cundo usar XGBoost?

- Cuando tienes datos en forma de **tabla** (como un archivo Excel o CSV).
- Cuando necesitas un modelo **preciso** para clasificar o predecir valores.
- Cuando ya probaste modelos ms simples (como regresin lineal) y quieres mejorar.


```python
# Importar libreras
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Cargar el dataset
iris = load_iris()
X = iris.data
y = iris.target

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)

# Crear y entrenar el modelo
model = XGBClassifier(
    n_estimators=100,         # Nmero de rboles (cuanto ms, ms preciso, pero ms lento)
    learning_rate=0.1,        # Qu tan rpido aprende el modelo (entre 0.01 y 0.3 suelen ir bien)
    max_depth=3,              # Qu tan profundo puede ser cada rbol (ms profundidad = ms complejidad)
    subsample=0.8,            # Porcentaje de datos que usa cada rbol (0.8 = 80%)
    colsample_bytree=0.8,     # Porcentaje de columnas que usa cada rbol
    reg_alpha=0.1,            # Regularizacin L1 (para evitar sobreajuste)
    reg_lambda=1.0,           # Regularizacin L2 (tambin ayuda a controlar la complejidad)
    use_label_encoder=False,  # Evita advertencia con etiquetas
    eval_metric='mlogloss'
)
model.fit(X_train, y_train)

# Predecir
y_pred = model.predict(X_test)

# Medir la precisin
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisin del modelo: {accuracy:.2f}")
```

### LightGBM

Framework de **gradient boosting** desarrollado por Microsoft que:  
 Construye rboles **verticalmente** (por hoja) vs horizontalmente (por nivel)  
 Optimizado para **grandes volmenes de datos**  
 Usa **GOSS** (Gradient-based One-Side Sampling) y **EFB** (Exclusive Feature Bundling)  

**Analoga**:  
"Como un arquitecto que construye un rascacielos piso por piso, no habitacin por habitacin"

##  **Ventajas Clave**  
| Caracterstica | Beneficio | Comparativa vs XGBoost |  
|----------------|-----------|------------------------|  
| **Crecimiento por hoja** | + Velocidad 3-5x ms rpida<br>+ Menor uso de memoria | XGBoost crece por niveles |  
| **Manejo eficiente de datos** | Soporta datasets >10GB en RAM limitada | Menos eficiente con big data |  
| **Soporte categrico nativo** | No necesita one-hot encoding | Requiere preprocesamiento |  
| **Paralelizacin ptima** | Entrenamiento ms rpido en multicore/GPU | Ms lento en paralelizacin |

##  **Cundo usarlo?**  
| Escenario | Ejemplo Prctico | Por qu LightGBM? |  
|-----------|------------------|--------------------|  
| **Grandes datasets** | 1M+ filas con 100+ caractersticas | Velocidad y eficiencia de memoria |  
| **Competiciones Kaggle** | Tabulares con mezcla numrica/categrica | Buen rendimiento out-of-the-box |  
| **Features categricas** | Datos de usuarios (pas, gnero, categoras) | Tratamiento nativo sin preprocesamiento |  
| **Prediccin en tiempo real** | Sistemas de recomendacin online | Inferencia ultra rpida |

##  **Parmetros Esenciales y sus Efectos**

### 1. **`num_leaves`** (Nmero de hojas por rbol)
| Accin | Efecto | Recomendacin |
|--------|--------|---------------|
| ** Aumentar** | + Mayor capacidad modelo<br>+ Captura patrones complejos<br>- Riesgo overfitting<br>- Ms memoria | Usar 30-50 para datasets medianos |
| ** Disminuir** | + Modelo ms simple<br>+ Menos overfitting<br>- Prdida de capacidad predictiva | No bajar de 10-15 en problemas complejos |

**Regla prctica**:  
`num_leaves  2^(max_depth)` pero con menos restricciones que en otros GBM

---

### 2. **`learning_rate`** (Tasa de aprendizaje)
| Accin | Efecto | Recomendacin |
|--------|--------|---------------|
| ** Aumentar** | + Convergencia ms rpida<br>- Riesgo de oscilaciones<br>- Puede no converger | Usar >0.1 solo para pruebas rpidas |
| ** Disminuir** | + Mejor optimizacin<br>+ Modelos ms precisos<br>- Requiere ms rboles | Valores tpicos: 0.01-0.3 |

**Relacin con `num_boost_round`**:  
`learning_rate` bajo + `num_boost_round` alto = mejor generalizacin

---

### 3. **`min_data_in_leaf`** (Mnimo datos por hoja)
| Accin | Efecto | Recomendacin |
|--------|--------|---------------|
| ** Aumentar** | + Reduce overfitting<br>+ Ms generalizable<br>- Prdida de patrones locales | Ideal 20-50 en datasets pequeos |
| ** Disminuir** | + Captura detalles finos<br>- Alto riesgo overfitting<br>- Hojas especficas | No usar <10 excepto en datasets muy grandes |

---

### 4. **`feature_fraction`** (Fraccin de features)
| Accin | Efecto | Recomendacin |
|--------|--------|---------------|
| ** Aumentar** | + Usa ms informacin<br>- rboles ms correlacionados | 1.0 para datasets con <50 features |
| ** Disminuir** | + Diversidad en rboles<br>+ Reduce overfitting<br>- Prdida informacin | 0.7-0.9 para datasets con 100+ features |

---

### 5. **`bagging_fraction`** (Submuestreo de datos)
| Accin | Efecto | Recomendacin |
|--------|--------|---------------|
| ** Aumentar** | + Cada rbol ve ms datos<br>- Menos diversidad | Usar con `bagging_freq` >0 |
| ** Disminuir** | + Mayor diversidad<br>+ Robustez<br>- Ms varianza | 0.8 es buen punto de partida |

---

### 6. **`lambda_l1`/`lambda_l2`** (Regularizacin)
| Accin | Efecto | Recomendacin |
|--------|--------|---------------|
| ** Aumentar** | + Modelo ms conservador<br>+ Reduce overfitting<br>- Posible underfitting | Comenzar con 0.1-1.0 |
| ** Disminuir** | + Modelo ms flexible<br>- Riesgo overfitting | Usar 0 para mxima flexibilidad |

---

##  **Parmetros Secundarios Clave**

### 1. **`max_depth`** (Profundidad mxima)
| Accin | Efecto | Recomendacin |
|--------|--------|---------------|
| ** Aumentar** | + rboles ms complejos<br>- Riesgo overfitting | LightGBM crece por hojas, no por profundidad |
| ** Disminuir** | + Modelo ms simple<br>+ Entrenamiento ms rpido | Usar -1 (sin lmite) en la mayora de casos |

### 2. **`cat_smooth`** (Suavizado categrico)
| Accin | Efecto | Recomendacin |
|--------|--------|---------------|
| ** Aumentar** | + Menos overfit en categoras raras<br>- Prdida de informacin | 10-100 para categoras con pocos ejemplos |
| ** Disminuir** | + Respuesta ms precisa<br>- Riesgo overfit | 1 para datasets grandes |

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns
from matplotlib.lines import Line2D

# 1. Visualizacin de Predicciones vs. Valores Reales en espacio reducido
def plot_results(X, y_true, y_pred, method='pca'):
    plt.figure(figsize=(16, 8))

    if method == 'pca':
        reducer = PCA(n_components=2, random_state=42)
        X_reduced = reducer.fit_transform(X)
        title_suffix = 'PCA'
    else:
        reducer = TSNE(n_components=2, random_state=42, perplexity=min(30, X.shape[0]//3))
        X_reduced = reducer.fit_transform(X)
        title_suffix = 't-SNE'

    # Crear subplots
    ax1 = plt.subplot(121)
    ax2 = plt.subplot(122)

    # Mapeo de colores para clases
    class_colors = {
        0: '#3498db',  # Azul
        1: '#2ecc71',  # Verde
        2: '#e74c3c'   # Rojo
    }

    # Grfico 1: Valores reales
    for cls in np.unique(y_true):
        idx = np.where(y_true == cls)
        ax1.scatter(X_reduced[idx, 0], X_reduced[idx, 1],
                   color=class_colors[cls],
                   s=70, alpha=0.7,
                   label=f'{wine.target_names[cls]} (Real)')

    ax1.set_title(f'Distribucin Real de Clases ({title_suffix})', fontsize=14)
    ax1.set_xlabel('Componente 1', fontsize=12)
    ax1.set_ylabel('Componente 2', fontsize=12)
    ax1.legend(loc='best')
    ax1.grid(alpha=0.2)

    # Grfico 2: Predicciones del modelo
    correct_mask = (y_true == y_pred)
    incorrect_mask = (y_true != y_pred)

    # Predicciones correctas
    for cls in np.unique(y_true):
        idx = np.where((y_pred == cls) & correct_mask)
        ax2.scatter(X_reduced[idx, 0], X_reduced[idx, 1],
                   color=class_colors[cls],
                   s=70, alpha=0.7,
                   label=f'{wine.target_names[cls]} (Correcto)')

    # Predicciones incorrectas (si las hay)
    if np.any(incorrect_mask):
        ax2.scatter(X_reduced[incorrect_mask, 0], X_reduced[incorrect_mask, 1],
                   color='black', marker='X',
                   s=150, alpha=1.0,
                   label='Error')

    ax2.set_title(f'Predicciones del Modelo ({title_suffix})', fontsize=14)
    ax2.set_xlabel('Componente 1', fontsize=12)
    ax2.legend(loc='best')
    ax2.grid(alpha=0.2)

    # Aadir porcentaje de precisin global
    accuracy = np.mean(correct_mask) * 100
    plt.suptitle(f'Precisin del Modelo: {accuracy:.2f}%', fontsize=16, y=0.95)

    plt.tight_layout()
    plt.savefig(f'results_comparison_{method}.png', dpi=300, bbox_inches='tight')
    plt.show()

# 2. Visualizacin de Probabilidades por Clase
def plot_class_probabilities(y_proba, y_true, class_names):
    plt.figure(figsize=(14, 8))

    # Preparar datos
    n_samples = len(y_true)
    sample_ids = np.arange(n_samples)

    # Crear grfico de barras apiladas
    bottom = np.zeros(n_samples)
    colors = ['#3498db', '#2ecc71', '#e74c3c']

    for i in range(len(class_names)):
        plt.bar(sample_ids, y_proba[:, i],
               bottom=bottom,
               color=colors[i],
               width=1.0,
               label=class_names[i])
        bottom += y_proba[:, i]

    # Aadir marcadores para la clase real
    for i, true_class in enumerate(y_true):
        plt.plot(i, 1.05, 'o', color=colors[true_class], markersize=8)

    # Personalizar grfico
    plt.title('Probabilidades de Prediccin por Muestra', fontsize=16)
    plt.xlabel('Muestras', fontsize=14)
    plt.ylabel('Probabilidad', fontsize=14)
    plt.ylim(0, 1.2)
    plt.grid(axis='y', alpha=0.3)

    # Crear leyenda personalizada
    legend_elements = [
        plt.Rectangle((0,0), 1, 1, color=colors[i], label=f'Prob. {name}')
        for i, name in enumerate(class_names)
    ]
    legend_elements += [
        Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[i],
              markersize=10, label=f'Clase Real: {name}')
        for i, name in enumerate(class_names)
    ]

    plt.legend(handles=legend_elements, loc='upper center',
              bbox_to_anchor=(0.5, -0.12), ncol=3, fontsize=12)

    # Aadir anotacin de precisin
    accuracy = np.mean(np.argmax(y_proba, axis=1) == y_true) * 100
    plt.figtext(0.5, 0.01, f'Precisin Global: {accuracy:.2f}%',
               ha='center', fontsize=12, bbox=dict(facecolor='#f8f9fa', alpha=0.8))

    plt.tight_layout(rect=[0, 0.05, 1, 0.95])
    plt.savefig('class_probabilities.png', dpi=300, bbox_inches='tight')
    plt.show()

# 3. Matriz de Confusin con Anlisis Detallado
def plot_enhanced_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=class_names,
               yticklabels=class_names,
               cbar=False)

    # Personalizar
    plt.title('Matriz de Confusin', fontsize=16, pad=20)
    plt.xlabel('Predicciones', fontsize=14, labelpad=15)
    plt.ylabel('Valores Reales', fontsize=14, labelpad=15)
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.yticks(rotation=0, fontsize=12)

    # Aadir mtricas por clase
    for i in range(len(class_names)):
        precision = cm[i, i] / cm[:, i].sum() if cm[:, i].sum() > 0 else 0
        recall = cm[i, i] / cm[i, :].sum() if cm[i, :].sum() > 0 else 0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        plt.text(len(class_names) + 0.5, i + 0.3,
                f'Precisin: {precision:.2%}\nRecall: {recall:.2%}\nF1: {f1:.2%}',
                ha='left', va='center', fontsize=11,
                bbox=dict(facecolor='white', alpha=0.8))

    # Aadir precisin global
    accuracy = np.sum(np.diag(cm)) / np.sum(cm)
    plt.figtext(0.5, 0.01, f'Precisin Global: {accuracy:.2%}',
               ha='center', fontsize=12,
               bbox=dict(facecolor='#f8f9fa', alpha=0.8))

    plt.tight_layout(rect=[0, 0.05, 1, 0.95])
    plt.savefig('enhanced_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

# --- EJECUCIN PRINCIPAL ---
# Obtener predicciones y probabilidades
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)

# 1. Visualizacin en espacio reducido (PCA y t-SNE)
print("Visualizando resultados con PCA...")
plot_results(X_test, y_test, y_pred, method='pca')

print("Visualizando resultados con t-SNE...")
plot_results(X_test, y_test, y_pred, method='tsne')

# 2. Visualizacin de probabilidades
print("Visualizando probabilidades de clase...")
plot_class_probabilities(y_proba, y_test, wine.target_names)

# 3. Matriz de confusin mejorada
print("Visualizando matriz de confusin detallada...")
plot_enhanced_confusion_matrix(y_test, y_pred, wine.target_names)

print("Visualizaciones completadas y guardadas!")
```
